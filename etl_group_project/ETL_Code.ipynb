```python
import pandas as pd
from sqlalchemy import create_engine
```

```python
# Import blockbuster csv file
blockbuster_file_csv = "Resources/blockbuster.csv"
blockbuster_df = pd.read_csv(blockbuster_file_csv)
blockbuster_df.head()
```

```python
blockbuster_df.drop_duplicates("title", inplace=True)
blockbuster_df.head()
```

```python
clean_blockbuster = blockbuster_df[["title", "Main_Genre", "rating", "year", "studio", "worldwide_gross"]]
```

```python
clean_blockbuster_df = pd.DataFrame(clean_blockbuster)
clean_blockbuster_df.head()
```

```python
clean_blockbuster_df.rename(columns={"Main_Genre":"genre", "year":"release_year"}, inplace=True)
clean_blockbuster_df.head()
```

```python
blockbuster = clean_blockbuster_df
```

```python
blockbuster_df = pd.DataFrame(blockbuster)
blockbuster_df.head()
```

```python
blockbuster_df.set_index("title", inplace=True)
blockbuster_df.head()
```

```python
# Import imdb csv file and create dataframe
imdb_file_csv = "Resources/IMDB-Movie-Data.csv"
imdb_data_df = pd.read_csv(imdb_file_csv)
imdb_data_df.head()
```

```python
# Drop duplicate titles
imdb_data_df.drop_duplicates("Title", inplace=True)
imdb_data_df.head()
```

```python
clean_imdb_data = imdb_data_df[["Title", "Director", "Actors", "Rating", "Votes", "Revenue (Millions)", "Metascore"]]
```

```python
clean_imdb_data_df = pd.DataFrame(clean_imdb_data)
clean_imdb_data_df.head()
```

```python
clean_imdb_data_df.rename(columns={"Title":"title", "Director":"director", "Actors":"actors", "Rating":"rating", "votes":"votes", "Revenue (Millions)":"revenue", "Metascore":"metascore"}, inplace=True)
clean_imdb_data_df.head()
```

```python
imdb = clean_imdb_data_df
```

```python
imdb_df = pd.DataFrame(imdb)
imdb_df.head()
```

```python
imdb_df.set_index("title", inplace=True)
imdb_df.head()
```

```python
# Import tmdb csv file and create dataframe
tmdb_file_csv = "Resources/tmdb_5000_movies.csv"
tmdb_data_df = pd.read_csv(tmdb_file_csv)
tmdb_data_df.head()
```

```python
# Drop duplicate titles
#tmdb_data_df.drop_duplicates("title", inplace=True)
tmdb_data_df.head()
```

```python
tmdb_data = new_tmdb_df
```

```python
tmdb_df = pd.DataFrame(tmdb_data)
tmdb_df.head()
```
